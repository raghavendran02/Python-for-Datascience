```python
import mysql.connector as conn
```

```python
## connecting to the database using 'connect()' method
db = conn.connect(
    host = "localhost",
    user = "root",
    passwd = "root",
    use_pure = True
)

print(db)
```

```python
try:
    cursor = db.cursor()
    cursor.execute("drop database InfoDeck")
except:
    print("unable to drop database")
```

```python
try:
    ## creating an instance of 'cursor' class which is used to execute the 'SQL' statements in 'Python'
    cursor = db.cursor()
    ## creating a databse called 'InfoDeck'
    ## 'execute()' method is used to compile a 'SQL' statement
    ## below statement is used to create tha 'InfoDeck' database
    cursor.execute("CREATE DATABASE InfoDeck")
except:
    print("Unable to create a database")
```

```python
#cursor = db.cursor()

## executing the statement using 'execute()' method
cursor.execute("SHOW DATABASES")

## 'fetchall()' method fetches all the rows from the last executed statement
databases = cursor.fetchall() ## it returns a list of all databases present

## printing the list of databases
print(databases)
```

```python
try:
    ## creating a table called 'users' in the â€˜InfoDeckâ€™ database
    cursor.execute("use InfoDeck")
    cursor.execute("CREATE TABLE users (id INTEGER, name VARCHAR(255), user_name VARCHAR(255))")
except:
    print("unable to create a table")
```

```python
## getting all the tables which are present in 'datacamp' database
cursor.execute("SHOW TABLES")

tables = cursor.fetchall() ## it returns list of tables present in the database

## showing all the tables one by one
for table in tables:
    print(table)
```

```python
query = "INSERT INTO users (id,name, user_name) VALUES (%s ,%s, %s)"
## storing values in a variable
#values = (1,"John", "John_Smith")
values = [(1,"John", "John_Smith"),(2,"Abc","ab_c"),(3,"Raghu","Raghu.n")]
## executing the query with values
count = 0
for row in values:
    #cursor.execute(query, values)
    cursor.execute(query, row)
    count += cursor.rowcount

print(count, "record inserted")
## to make final output we have to run the 'commit()' method of the database object
db.commit()
```

```python
#cursor = db.cursor()
query = "SELECT * FROM users"

## getting records from the table
cursor.execute(query)


## fetching all records from the 'cursor' object
records = cursor.fetchall()
```

```python
for record in records:
    print(record)
```
