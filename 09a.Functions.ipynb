```python
def replace_negative(val):
    if val<0:
        out=0
    else:
        out=val
    return out
```

```python
replace_negative(-5)
```

```python
replace_negative(10)
```

```python
lst = [10,20,-5,7,9,-7,-6,-90,100]
#Clean the list such that the -ve numbers are replaced by 0
lst_clean = []
for num in lst:
    val = replace_negative(num)
    lst_clean.append(val)
lst_clean
```

```python
def greaterThanAvg(lst):
    def avg():
        sum = 0
        for num in lst:
            sum+=num
        return(sum/len(lst))
    average = avg()
    lst1 = []
    for num in lst:
        if num > average:
            lst1.append(num)
    return lst1
```

```python
lst = [15,20,30,40,60,10]
greaterThanAvg(lst)
```
