```python
import sqlite3
```

```python
database = "student.db"
```

```python
drop_table = "drop table student"
try:
    with sqlite3.connect(database) as conn:
        cursor=conn.cursor()
        cursor.execute(drop_table)
        conn.commit()
except:
    print("unable to drop the table")
```

```python
#Creating a table
create_table = "create table student(studentid integer primary key, name text not null, branch text)"
```

```python
try:
    with sqlite3.connect(database) as conn:
        cursor = conn.cursor()
        cursor.execute(create_table)   
        conn.commit()

except sqlite3.OperationalError as e:
    print(e)
```

```python
try:
    with sqlite3.connect(database) as conn:
        cursor = conn.cursor()
        cursor.execute("insert into student values(101,'Ram','CS')")
        cursor.execute("insert into student values(102,'Mohan','IS')")
        cursor.execute("insert into student values(103,'Priya','IS')")
        cursor.execute("insert into student values(104,'Surya','EC')")
        conn.commit()
except sqlite3.OperationalError as e:
    print(e)
```

```python
conn = sqlite3.connect('attrition.db')
data.to_sql('emp_attr', conn, if_exists='replace', index=False)
```

```python
pd.read_sql('select * from emp_attr limit 10', conn)
```

```python
pd.read_sql('select count(*) as "No of Records" from emp_attr', conn)
```

```python
query = 'select Department,count(*) as "No of emp", sum(dailyrate) as "total daily rate" from emp_attr group by Department order by 2 desc'
```

```python
pd.read_sql(query, conn)
```

```python
# Department wise total attrition
```

```python
query = 'select Department,sum(attrition) as "Total attrition" from emp_attr group by Department order by 2 desc'
```

```python
out1 = pd.read_sql(query, conn)
out1
```
